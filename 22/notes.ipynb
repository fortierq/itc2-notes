```python
import pandas as pd
pd.options.display.float_format = '{:,.1f}'.format

df_ds2 = pd.read_excel("notes_itc.xlsx", sheet_name="ds2")
df_id = pd.read_excel("notes_itc.xlsx", sheet_name="id")
df_id["nom"] = df_id["nom"].apply(lambda x: x.lower())
df_ds2 = pd.merge(df_id, df_ds2, on=["nom", "classe"], how="right")
df_ds2["brut"] = (df_ds2[range(1, 18)]*(df_ds2.query("nom == 'bareme'").iloc[0][range(1, 18)])/5).sum(axis=1)
```

```python
df_classe = df_ds2[["classe", "brut"]].groupby("classe").agg(["mean", "std"])["brut"]
df_classe["moyenne"] = {"mpp": 13, "pcc": 11}
df_classe["a"] = 3.5/df_classe["std"]
df_classe["b"] = df_classe["moyenne"] - df_classe["a"]*df_classe["mean"]
classes = df_ds2["classe"].dropna()
df_ds2.loc[1:, "note"] = df_ds2.loc[1:, "brut"]*df_classe.loc[classes, "a"].values + df_classe.loc[classes, "b"].values
df_ds2 = df_ds2.round(1)
df_ds2["rang"] = 0
df_ds2.loc[1:, "rang"] = df_ds2.loc[1:].groupby("classe")["note"].rank("min", ascending=False)
df_ds2.loc[0, "rang"] = 0
df_anonyme = df_ds2.drop(columns=["nom", "prenom", "brut"])
df_anonyme.to_csv("ds2.csv", index=False)
df_ds2
```
