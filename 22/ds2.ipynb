```python
import pandas as pd
pd.options.display.float_format = '{:,.1f}'.format

df_ds2 = pd.read_csv("ds2.csv")
df_ds2 = df_ds2.iloc[1:, :]
questions = list(map(str, range(1, 18))) + ["classe"]
df_questions = df_ds2[questions]
```

```python
n_eleves = len(df_questions) - 1
((n_eleves - df_questions.drop(columns="classe").isnull().sum())/n_eleves).plot.bar(title="Pourcentage de réponses par question")
```

```python
df_questions.groupby("classe").mean().T.plot.bar(figsize=(10, 5), title="Taux de réussite par question")
```

```python
df_questions = df_ds2.copy()
df_questions.loc[df_questions.index, questions] = (df_questions[questions]*df_questions.iloc[0]/5)
df_questions.loc[1:, questions + ["classe"]].groupby("classe").mean().T.plot.bar(figsize=(10, 5), title="Moyenne par question")
# df_questions
```

```python
import matplotlib.pyplot as plt

for c in df_ds2["classe"].unique():
    df_ds2.query(f"classe == '{c}'")[["note", "rang"]].set_index("rang").sort_values(by="note", ascending=False).plot.bar(title=f"Classe {c}")
    plt.show()
```
